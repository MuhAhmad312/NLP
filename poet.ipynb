```python
import random 
import nltk 
from nltk.tokenize import word_tokenize, sent_tokenize 
from nltk.util import ngrams
```

```python
with open('ghalib.txt','r', encoding='utf-8') as file1: 
    txt1 = file1.read().split('\n')
with open('iqbal.txt', 'r', encoding='utf-8') as file2: 
    txt2 = file2.read().split('\n') 
```

```python
txt3 = txt1 + txt2 
txt3 = " ".join(txt3)
txt3
```

```python
words = word_tokenize(txt3)
bigrams = list(ngrams(words,2))
print(bigrams)
```

```python
bi_cfd = nltk.ConditionalFreqDist(bigrams)
for i in range(3): # 3 paragraphs
    for j in range(4): # 4 verses in each paragrapg
        verse = [] 
        for k in range(random.randint(7,11)): 
            if not verse: 
                word = random.choice(words) # randomly choose the first word
            else: 
                word = bi_cfd[verse[-1]].max() # using conditional frequencies find the next word 
            verse.append(word) 
        print(' '.join(verse)) 
    print()
```
